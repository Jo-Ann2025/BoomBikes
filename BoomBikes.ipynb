```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# Example: assuming X_train and y_train are already loaded
# Step 1: Check data types of X_train and y_train
print(X_train.dtypes)
print(y_train.dtypes)

# Step 2: Convert non-numeric columns in X_train to numeric (if any)
# For example, convert categorical columns to numeric using one-hot encoding or label encoding

# If there are categorical columns, you can use pd.get_dummies() to convert them
X_train = pd.get_dummies(X_train, drop_first=True)

# Step 3: Ensure y_train is numeric (if it's not, convert it)
y_train = pd.to_numeric(y_train, errors='coerce')

# Step 4: Handle missing values in X_train and y_train
# If there are missing values, either drop them or fill with a value (mean, median, etc.)

# Drop rows where there are missing values in either X_train or y_train
X_train = X_train.dropna()
y_train = y_train[X_train.index]  # Align y_train with X_train after dropping rows

# Step 5: Add constant (intercept term) to X_train
X_train_sm = sm.add_constant(X_train)

# Step 6: Fit the OLS model
lr_obj = sm.OLS(y_train, X_train_sm)
lr = lr_obj.fit()

# Step 7: Display the results
print(lr.summary())
```

```python
season_map = {
    1: 'Spring', 
    2: 'Summer', 
    3: 'Fall', 
    4: 'Winter'
}
```

```python
yr_map = {
    0: '2018', 
    1: '2019', 
}
```

```python
mnth_map = {
    1: 'January', 
    2: 'February', 
    3: 'March', 
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December'
}
```

```python
holiday_map = {
    0: 'weather day is not holiday', 
    1: 'weather day is a holiday', 
}
```

```python
weekday_map = {
    1: 'Monday', 
    2: 'Tuesday', 
    3: 'Wednesday', 
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}
```

```python
workingday_map = {
    0: 'working day', 
    1: 'weekend or holiday', 
}
```

```python
weathersit_map = {
    1: 'Clear, Few clouds, Partly cloudy, Partly cloudy', 
    2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist', 
    3: 'List Snow, Light Rain + Thunderstorn + Scattered clouds, Light Rain + Scattered clouds', 
    4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'
}
```

```python
df['season'] = df['season'].map(season_map)
df['yr'] = df['yr'].map(yr_map)
df['mnth'] = df['mnth'].map(mnth_map)
df['holiday'] = df['holiday'].map(holiday_map)
df['weekday'] = df['weekday'].map(weekday_map)
df['workingday'] = df['workingday'].map(workingday_map)
df['weathersit'] = df['weathersit'].map(weathersit_map)
```
